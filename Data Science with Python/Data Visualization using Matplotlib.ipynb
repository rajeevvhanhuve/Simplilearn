### Pie Chart
Pie chart shows distribution of data based on proportion of pie occupied. 

#### The explode property
To highlight a particular wedge of the pie chart, we use explode to separate it from the rest of the chart. 


```python
t_mov= pd.read_csv(r'datasets/types_movies.csv')
t_mov.head()
```

```python
t_mov['Sector']
```

```python
plt.pie(
#     t_mov['Percentage'],        # the data will be provided
    [7000,15000,3000,4000,1000], 
    labels=t_mov['Sector'],      # the label names to be provided
    autopct='%.1f%%')            # it converts the numbers to % format directly, here .1f means - float with 1 decimal

plt.show()
```

```python
t_mov.head()
```

```python

color_list = ['orange', 'brown', 'blue',  'green',   'yellow']

explode_val= (   0.1,       0,      0.2,      0,        0.1) # it lies in the range 0 to 1

#            ['Sci-Fi',  'Drama', 'Action','Comedy', 'Romance']

plt.pie(t_mov['Percentage'], 
        labels=t_mov['Sector'], 
        autopct='%.1f%%',  # convert the numbers to % format where .1f represent numbers to 1 decimal digits
        colors=color_list, 
        explode=explode_val
       )

plt.show()
```

## Boxplot & Violinplot

```python
# Box plot is used to detect outlier in the data

# Age =[-1,0,30,35,50,110,150,300]


# Methods of Identifying Outlier

# 1) any value below 5 Percentile and above 95 percentile is considered as outlier
# 2) Any value beyond 2 STD to the left and right of the Mean value is considered as outlier in the data/column.
# 3) Using IQR Method - Inter Quantile range

# How to identify the outliers

# IQR=Q3-Q1 (75% - 25%)


# Minor Fencing

# lower_fence = Q1-(1.5 * IQR)  ## ANY VALUE BELOW LOWER_FENCE WOULD BE AN OUTLIER e.g -- 10

# Upper_Fence = Q3 + (1.5 * IQR) ## ANY VALUE ABOVE UPPER_FENCE WOULD BE AN OUTLIER e.g. -- 120
 
# Age =[-1,0,30,35,50,110,150,300]

# outliers = -1, 0, 150, 300
```

# Barplot

```python
countries=pd.read_csv('datasets/countries.csv')
countries.head(2)
```

```python
data_2007=countries[countries['year']==2007]
top_10_p=data_2007.sort_values('population',ascending=False).head(10)
top_10_p.head()
```

```python
plt.figure(figsize=(12,6))
plt.bar(top_10_p['country'],top_10_p['population']/10**9)  # 10**9 = 1 Billion
plt.xlabel('Countries')
plt.ylabel('Population in Billions')
plt.title('10 Most Populous Countries in 2007')
plt.show()
```

# Scatterplot

```python
# plt.scatter(x,y)

plt.scatter(data_2007['gdpPerCapita'],data_2007['lifeExpectancy'])

plt.title('Life Expectancy vs GDP Per Capita in 2007 ')
plt.xlabel('GDP per Capita($)')
plt.ylabel('Life Expectancy')
plt.show()
```
